```python
from sklearn.preprocessing import StandardScaler
df3 = pd.read_csv('transaction.csv', delimiter=",")
con_vars = ['payerAccountNumber','originalAmount', 'payeeIdTransformed',
           'payerIdTransformed']
scaler = StandardScaler()
df3[con_vars]=scaler.fit_transform(df3[con_vars])
df3.head(5)
df3.to_csv('transaction_standerdize.csv',index=False)
```
